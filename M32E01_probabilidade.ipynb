1. Uma máquina embala ração para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatório, com distribuição Normal, com média 15kg, mas com desvio padrão 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produção tenha mais que 15,2 kg.

```python
import numpy as np
from scipy.stats import norm

# Parâmetros da distribuição
mu = 15
sigma = 0.1

# Valor de interesse
x = 15.2

# Calcula a probabilidade de P(X > 15.2)
probabilidade = 1 - norm.cdf(x, loc=mu, scale=sigma)

print(f"Probabilidade de um pacote pesar mais que 15,2 kg: {probabilidade:.4f}")
```

2. Da máquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
from scipy.stats import norm

# Parâmetros da distribuição
mu = 15
sigma = 0.1

# Percentil de 95%
percentil_95 = norm.ppf(0.95, loc=mu, scale=sigma)

print(f"Percentil de 95%: {percentil_95:.4f} kg")
```

4. Faça uma visualização gráfica dos dados do item anterior.

```python
import numpy as np
from scipy.stats import norm, t
import matplotlib.pyplot as plt

# Valores de Y
Y = np.linspace(-6, 6, 200)

# CDF da Normal padrão
cdf_normal = norm.cdf(Y, loc=0, scale=1)

# CDF da t de Student com 5 graus de liberdade
cdf_tstudent = t.cdf(Y, df=5)

# Plotando os resultados
plt.figure(figsize=(10, 6))
plt.plot(Y, cdf_normal, label='Normal Padrão', color='blue')
plt.plot(Y, cdf_tstudent, label='t-Student (df=5)', color='red', linestyle='--')
plt.title('Função de Distribuição Acumulada (CDF)')
plt.xlabel('Y')
plt.ylabel('P(X < Y)')
plt.grid(True)
plt.legend()
plt.show()
```

5. 3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
import numpy as np
from scipy.stats import norm, t

# Gerar 20 valores de probabilidade entre 0 e 1 (excluindo extremos)
Y = np.linspace(0.01, 0.99, 20)

# Quantis para Normal padrão
quantis_normal = norm.ppf(Y, loc=0, scale=1)

# Quantis para t-Student com 5 graus de liberdade
quantis_t = t.ppf(Y, df=5)

# Mostrar resultados
for p, qn, qt in zip(Y, quantis_normal, quantis_t):
    print(f"P = {p:.2f} | Normal padrão quantil = {qn:.4f} | t-Student(5) quantil = {qt:.4f}")

```

6. Faça um gráfico com os dados do item anterior.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t

# Probabilidades (evitando 0 e 1 para evitar ±inf)
Y = np.linspace(0.01, 0.99, 20)

# Cálculo dos quantis
quantis_normal = norm.ppf(Y, loc=0, scale=1)
quantis_t = t.ppf(Y, df=5)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(Y, quantis_normal, label='Normal Padrão', marker='o')
plt.plot(Y, quantis_t, label='t-Student (df=5)', marker='s')

plt.title('Quantis para Normal Padrão e t-Student (df=5)')
plt.xlabel('Probabilidade (p)')
plt.ylabel('Quantil (inversa da CDF)')
plt.legend()
plt.grid(True)
plt.show()
```
